2015-2016

弹性网

```python
from sklearn.metrics import r2_score
```

```python

clf5 = ElasticNet(alpha=3, l1_ratio=0.5, fit_intercept = True, max_iter = 2000,tol = 0.1)
clf5.fit(x_train5, y_train5)
predict_y5 = clf5.predict(x_test5)

df5 = pd.DataFrame(data=predict_y5, index=L5,columns=["predicted rate"])
```

```python
import statsmodels.api as sm
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
train5, test5 = predict_y5, X[size-52:size]
```

```python

mod5 = sm.tsa.SARIMAX(endog = test5,exog = train5, order=(1,1,1))
fit_res = mod5.fit(disp=0)
#print(fit_res.summary())
output5 = fit_res.predict()
ar5 = pd.DataFrame(data=output5, index=L5,columns=["predicted rate"])
error5 = mean_squared_error(test5, output5)
rmse5 = np.sqrt(error5)
```

单独ar

```python
train5, test5 = X[0:size-52], X[size-52:size]
history5 = [x for x in train5]
predictions5 = list()
for t in range(len(test5)):
    #model5 = ARIMA(history5, order=(3,1,1))
    model5 = sm.tsa.SARIMAX(endog = history5, order=(1,1,1))
    model_fit5 = model5.fit(disp=-1)
    output5 = model_fit5.forecast()
    yhat5 = output5[0]
    predictions5.append(yhat5)
    obs5 = test5[t]
    history5.append(obs5)
e5 = mean_squared_error(test5, predictions5)
r5 = np.sqrt(e5)
```

```python
predictions5 = pd.DataFrame(predictions5,index=L5,columns=['predictive_rate'])

plt.ylabel("ILI rate per 100,000 people")
plt.xlabel("test period from 2015 to 2016")

plt.plot(df5, label='Elastic Net',color='r')
plt.plot(ar5, label='SARIMA with Elastic net',color='g',linestyle='--')
plt.plot(predictions5, label='ARIMAX',color='b',linestyle='--')
plt.plot(data_test5, label='True ILI Rate')

#plt.title('RMSE of Elastic Net: %.4f'% np.sqrt(sum((predict_y5-y_test5)**2)/len(y_test5)))
#plt.title('RMSE of ARIMAX: %.4f'% r5)
#plt.suptitle('RMSE of ARIMAX + Elastic net : %.4f'% rmse5)
#plt.suptitle('RMSE of ARIMAX : %.4f'% r5 )
print ('RMSE of Elastic Net: %.4f'% np.sqrt(sum((predict_y5-y_test5)**2)/len(y_test5)))
print ('RMSE of ARIMAX + Elastic net: %.4f'% rmse5 )
print ('RMSE of ARIMAX : %.4f'% r5)
print ('mae of Elastic Net: %.4f'% mean_absolute_error(predict_y5,y_test5))
print ('mae of ARIMAX + Elastic net: %.4f'% mean_absolute_error(ar5['predicted rate'], y_test5))
print ('mae of ARIMAX : %.4f'% mean_absolute_error(predictions5, y_test5))

enet5_r2 = r2_score(predict_y5, y_test5)
arimaenet5_r2 = r2_score(ar5['predicted rate'], y_test5)
arima5_r2 = r2_score(predictions5, y_test5)
print('R2 of Enet: %.4f' % enet5_r2)
print('R2 of Enet+amrimax: %.4f' % arimaenet5_r2)
print('R2 of arima: %.4f' % arima5_r2)

plt.legend(['Elastic Net', 'ARIMAX + Elastic net', 'ARIMAX', 'True ILI Rate'], loc='upper left')
```
